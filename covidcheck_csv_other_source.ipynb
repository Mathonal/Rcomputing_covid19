```python
# # GENERAL INITIALISATION
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

filename = 'sursaudcorona_inpout.csv'
fullPath = './RAWDATA/'+filename

rawDF = pd.read_csv(fullPath,sep=';')
rawDF
```

```python
rawDF.columns
```

```python
# KEEP WANTED COLUMNS
#dep	integer	Department
#date_de_passage	string($date)	Date of notice
#sursaud_cl_age_corona	integer	Age group
#nbre_pass_corona	integer	Number of emergency room visits for suspicion of COVID-19 
#nbre_hospit_corona	integer	Number of hospitalizations among emergency department visits for suspicion of COVID-19 
columnlist = ['dep', 'date_de_passage', 'sursaud_cl_age_corona', 'nbre_pass_corona', 'nbre_hospit_corona']

workDF = rawDF[columnlist]
workDF
```

```python
frDF = workDF[['date_de_passage', 'nbre_pass_corona', 'nbre_hospit_corona']]
frDF = frDF.fillna(0)
frDF
```

```python
frDF.isna().sum()
```

```python
frDF = frDF.groupby('date_de_passage').sum()
frDF
```

```python
#small visualization of data across all one dataset
fig = plt.figure(figsize=(10, 5))
frDF.plot(subplots=True,figsize=(15, 10))
```
